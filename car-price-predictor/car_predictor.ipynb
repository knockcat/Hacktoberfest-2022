```python
#Indepndent and dependent features
X=final_dataset.iloc[:,1:]
y=final_dataset.iloc[:,0]
```

```python
#Feature Importance
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
#plotting graph of feature importance for better visualization
feature_importances = pd.Series(model.feature_importances_,index=X.columns)
feature_importances.nlargest(5).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
#rf_random = RandomForestRegressor()
```

```python
#Hyperparameters
import numpy as np

#Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start=100,stop=1200,num=12)]
print(n_estimators)
```

```python
#Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5,30,num=6)]
#Minimum no. of samples required to split a node
min_samples_split=[2,5,10,15,100]
#Minimum no. of samples required at each leaf node
min_samples_leaf = [1,2,5,10]
#Number of features to consider at every split
max_features = ['auto','sqrt']
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_grid = {'n_estimators':n_estimators,'max_depth':max_depth,'max_features':max_features,'min_samples_leaf':min_samples_leaf,'min_samples_split':min_samples_split}
print(random_grid)
```

```python
#Used the random grid to search for the best hyperparameters
#First creatnig the base model to tune
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator=rf,param_distributions=random_grid,scoring='neg_mean_squared_error',n_iter=10,cv=5,verbose=2,random_state=42,n_jobs=1)
```

```python
print(y_train.shape,X_train.shape)
```

```python
rf_random.fit(X_train,y_train)
```

```python
predictions = rf_random.predict(X_test)
```

```python
predictions
```

```python
sns.distplot(y_test-predictions)
```
